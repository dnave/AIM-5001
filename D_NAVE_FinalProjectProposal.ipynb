## Introduction: 
The entertainment industry has changed significantly since streaming companies have emerged. Today, it is very easy to stream a movie of one’s choice, simply by using one of many streaming services. Thus, viewers have many choices to evaluate their subscriptions according to the streaming service offerings. For example, as of January 2020, IMDb has approximately 6.5 million titles (including episodes) and 10.4 million personalities in its database, as well as 83 million registered users. However, as rating systems vary, it is confusing to tell which one is trustable. Oftentimes, blind worship of viewers’ ratings and review comments is misleading. In this project, we are going to explore the relationships between viewer’s ratings and movie characteristics, as well as the streaming services in which they are offered. The research results will reflect the viewer’s rating preferences and provide a reference for new audiences to rationally select movies based on ratings.

## Research Questions:
I. Do the factors in the dataset influence the ratings? (for example streaming service, directors, genres, countries, languages, producing years).<br>
Movie ratings may be affected by different aspects. The streaming services have different qualities, directors’ directing characteristics, and different topics under various genres, countries and languages are all factors that influence viewers' preferences. The main factors could be concluded through this question.


II. Are there any differences between the IMDb rating and the Rotten Tomatoes?<br>
IMDb rating and Rotten Tomatoes ratings are two of the most popular rating systems. However, the two systems are based on different rules, algorithm, and reviewer groups. The relationship between the two rating systems will be explored to provide general ideas for new audiences how to treat the two rating systems.


III. Is there a difference in movie ratings based on cultural reviewer groups? Will the reviewer groups with different cultural backgrounds affect the ratings as well?<br>
For this question, we are going to attempt to use web scraping method to get the data from a non-English speaking country website and compare the rating result with IMDb ratings. Cultural differences of the two countries may affect the ratings as well. Understanding of the cultural differences will be benefit to the objective and fair evaluation of movies.
