#### Research Narrative:
I am interested in exploring how happiness is related to geographical regions through a set of mediating variables. In order to do so,  I will draw the boxplots of regions and their corresponding total happiness Score. 

```python
# Measuring the Score over the Regions by using Seaborn Boxplot
# Define a new figure for plotting, defining  plot size
plt.figure(figsize = (15,7))

# Assigning values to x axis = Region and Y axis = Score, Organizing the order alphabetically
ax = sns.boxplot(x="Region", y="Score", data=WHR, width=0.5, order = ["Australia and New Zealand", 
                                                                      "Central and Eastern Europe", 
                                                                      "Eastern Asia", 
                                                                      "Latin America and Caribbean", 
                                                                      "Middle East and Northern Africa", 
                                                                      "North America", 
                                                                      "Southeastern Asia", 
                                                                      "Southern Asia", 
                                                                      "Sub-Saharan Africa", 
                                                                      "Western Europe"])
medians = WHR.groupby(['Region'])['Score'].median().round(2) 
vertical_offset = WHR['Score'].median() * -0.04

for xtick in ax.get_xticks():
    ax.text(xtick,medians[xtick] + vertical_offset,medians[xtick], 
            horizontalalignment='center',size='x-small',color='k',weight='semibold')


# Assigning visualization charactaristics to x axis
plt.xticks(rotation=45, horizontalalignment='right')

# Define tick size, alignment and rotation
plt.rcParams['xtick.labelsize']=10; 
plt.rcParams['ytick.labelsize']=10;

# Show the new figure
plt.show
```

```python
# Measuring the Score over the Regions by using Matplotlib Boxplot
# Define a new figure for plotting
fig = plt.figure() 

# Assigning values to columns = Region by the Scores
WHR.boxplot(column=["Score"], by = ["Region"], figsize = (15,7))

# Define tick size, alignment and rotation
plt.xticks(rotation=45, horizontalalignment='right')

# Show the new figure
plt.show
```

#### Observations:
First, we can see that the median varies between the regions, from a minimum of 4.13 in SubSaharan Africa to a maximum of 7.32 in Australia and New Zealand.  Furthermore, we see that the Middle East and North Africa region have the highest interquartile range and the highest variance between countries.  It is important to note that the low variance of Australia and New Zealand, and North America is due to a small sample size of two and three, respectively.  The boxplots show only three outliers for Latin America and Caribean region, Southern Africa region, and the Sub- Saharan Africa region which might indicate that countries within geographical regions have common happiness scores.
